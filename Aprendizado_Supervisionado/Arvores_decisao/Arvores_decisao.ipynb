#Exemplo de árvore de decisão

## Para classificação



```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

#importamos o conjunto de dados IRIS
from sklearn import datasets
iris = datasets.load_iris()


```

Construo meu conjunto de dados

```python
X = iris['data'][:,2:] #seleciono somente a largura e o comprimento de pétala como atributos
y = iris.target
print(y)

```

 Treinamos uma árvore de decisão

```python
tree_clf=DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X,y)

```

```python
# criando o fig e o axes
fig, ax = plt.subplots(figsize=(8,6),dpi=92)

#criando o plot
plot_tree(tree_clf,ax=ax)

#plotando o gráfico
plt.tight_layout();
```

Pode-se vizualizar a árvores  de decisão treinada usando o método export_graphviz()

```python
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(
    tree_clf,
    # out_file="iris_tree.dot",
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)

graph = graphviz.Source(dot_data)
graph
```

Fazendo predições

```python
tree_clf.predict_proba([[5, 1.5]]) #flor com pétala de 5 cm de comprimento e 1,5 cm de largura
#dá a probabilidade de pertencimento às diferentes classes
```

```python
tree_clf.predict([[5, 1.5]])
#retorna a classe mais provável
```

## Para regressão

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X,y)
```

```python
# criando o fig e o axes
fig, ax = plt.subplots(figsize=(8,6),dpi=92)

#criando o plot
plot_tree(tree_reg,ax=ax)

#plotando o gráfico
plt.tight_layout();
```

Fazendo predições

```python
tree_reg.predict([[5, 1.5]])
```
